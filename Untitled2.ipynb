````python
def normalize(x):

    if x is None:
        return ""

    if isinstance(x, list):
        return " ".join(map(str, x))

    return str(x)
````

````python
for scheme in schemes:

    scheme["search_text"] = " ".join([

        normalize(scheme.get("name")),
        normalize(scheme.get("description")),
        normalize(scheme.get("eligibility")),
        normalize(scheme.get("benefits")),
        normalize(scheme.get("funding")),
        normalize(scheme.get("keywords"))

    ]).lower()
````

````python
import re

def search_schemes(query, schemes, top_k=5):

    words = re.findall(r"\w+", query.lower())

    scored = []

    for scheme in schemes:

        score = 0

        for word in words:

            if word in scheme["search_text"]:
                score += 1

        if score > 0:
            scored.append((score, scheme))

    scored.sort(key=lambda x: x[0], reverse=True)

    return [item for score, item in scored[:top_k]]
````

````python
query = "Which schemes provide grants for biotech startups?"

results = search_schemes(query, schemes)

for scheme in results:
    print(scheme["name"])
````

````python
import json

context = json.dumps(results, indent=2)
````
